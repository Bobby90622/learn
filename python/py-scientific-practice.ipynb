```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt

import matplotlib as mpl

mpl.rcParams['figure.dpi'] = 300
```

```python
gdp = pd.read_csv('assets/gdp.csv', skiprows=4)
gdp.head(10)
```

```python
df = gdp.query('`Country Name` in ["China", "Japan", "United States"]').transpose()
df.columns = df.iloc[0]
df = df[4:-2]
df = df.applymap(lambda x: x / 1e12)
df
```

```python
from datetime import datetime
import numpy as np
from scipy import stats

from statsmodels.graphics.api import qqplot
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima_model import ARMA
```

```python
def stationarity_test(ts):
    rolmean = ts.rolling(4).mean()
    rolstd = ts.rolling(4).std()
    
    plt.figure(figsize=(12, 4))
    orig = plt.plot(ts, color='blue', label='Original')
    mean = plt.plot(rolmean, color='red', label='Rolling Mean')
    std = plt.plot(rolstd, color='black', label='Rolling Std')
    
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show(block=False)
    
    print('Results of Dickey-Fuller Test:')
    dftest = adfuller(ts, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statostic', 'p-value', '# Lags Used', 'Number of Observations Used'])
    
    for key, value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
        
    print(dfoutput)
```

```python
cn_gdp = df['China']
cn_gdp.index = pd.to_datetime(cn_gdp.index)
cn_gdp.head(10)
```

```python
stationarity_test(cn_gdp)
```

```python
cn_gdp_lg = np.log(cn_gdp)

stationarity_test(cn_gdp_lg)
```

```python
cn_gdp_diff1 = cn_gdp_lg.diff(1)
cn_gdp_diff1.dropna(inplace=True)
stationarity_test(cn_gdp_diff1)
```

```python
plot_acf(cn_gdp_diff1, lags=15).set_size_inches(3, 3)
plot_pacf(cn_gdp_diff1, lags=15).set_size_inches(3, 3)
plt.show()
```

```python
model10 = ARMA(cn_gdp_diff1, (1, 0)).fit()
```

```python
resid = model10.resid

fig = plt.figure(figsize=(4, 4))
ax = fig.add_subplot(111)
fig = qqplot(resid, line='q', ax=ax, fit=True)
plt.show()
```

```python
import statsmodels.api as sm
print(sm.stats.durbin_watson(resid.value))
```

```python
len(cn_gdp_diff1)
```

```python
cn_arma_pred = model10.predict(start=0, end=60)
```

```python
plt.figure(figsize=(12, 4))
orig = plt.plot(cn_gdp_diff1, color='blue', label='Original')
predict = plt.plot(cn_arma_pred, color='red', label='Predict')
plt.legend(loc='best')
plt.title('Original & Predict')
plt.show(block=False)
```

```python
# reverse diff
cn_pred_log = cn_arma_pred.add(cn_gdp_lg.shift(1))
cn_pred_log[-1] = cn_arma_pred[-1] + cn_gdp_lg[-1]

# reverse log
cn_pred = np.exp(cn_pred_log)
cn_pred.dropna(inplace=True)
```

```python
plt.figure(figsize=(12, 4))
orig = plt.plot(cn_gdp, color='blue', label='Original')
predict = plt.plot(cn_pred, color='red', label='Predict')
plt.legend(loc='best')
plt.title('Original & Predict')
plt.show(block=False)
```

```python
error_rate = (cn_pred - cn_gdp) / cn_gdp
plt.figure(figsize=(12, 3))
plt.scatter(error_rate.index, error_rate, color='r')
plt.grid(True)
plt.figure(figsize=(3, 3))
plt.hist(error_rate, bins=20)
plt.grid(True)
plt.show()
```
